#### Use the appropriate plot to visualize the column `Gender`. There are 2 ways of doing it. Do it both ways.

*Hint: Use matplotlib and seaborn*

```python
# your code here- 1st way
male = titanic.loc[titanic['Gender'] == 'male'].count()[0]
female = titanic.loc[titanic['Gender'] == 'female'].count()[0]

plt.figure(figsize=(6,3))

labels = ['male', 'female']
colors = ['g', 'y']

plt.pie([male, female], labels = labels, colors=colors, autopct='%.2f %%')

plt.title('Gender % in the Titanic')

plt.show()
```

```python
# your code here- 2nd way
plt.figure(figsize=(6,3))
sns.countplot(titanic.Gender)
plt.show()
```
